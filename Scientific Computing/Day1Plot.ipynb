```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data to plot
x = np.linspace (0, 10, 100)
y = 4 + 2 * np.sin(2 * x)
```

```python
# plot the data
plt.plot (x,y,linewidth = 2.0)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# display the plot
plt.show()
```
