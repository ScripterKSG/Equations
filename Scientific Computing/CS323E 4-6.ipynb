```python
'''
F(x) = a_0 + a_1 cosx + a_2cos(2x) + a_3cos(3x) + b_1 sin(x) + b_2sin(2x) + b_3sin(3x)
F(x) = 1, 0 < x < \pi
F(x) = 0, \pi < x < 2\pi

First we try to find the coefficients
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate

a_0 = 1
a_1 = 1/np.pi * integrate.quad(lambda x: np.cos(x), 0, np.pi)[0]
a_2 = 1/np.pi * integrate.quad(lambda x: np.cos(2*x), 0, np.pi)[0]
a_3 = 1/np.pi * integrate.quad(lambda x: np.cos(3*x), 0, np.pi)[0]

b_0 = 0
b_1 = 1/np.pi * integrate.quad(lambda x: np.sin(x), 0, np.pi)[0]
b_2 = 1/np.pi * integrate.quad(lambda x: np.sin(2*x), 0, np.pi)[0]
b_3 = 1/np.pi * integrate.quad(lambda x: np.sin(3*x), 0, np.pi)[0]

x = np.linspace(0, 2*np.pi, 100)

print(a_1,a_2,a_3,b_1,b_2,b_3)
f = a_0 + a_1 * np.cos(x) + a_2 * np.cos(2*x) + a_3 * np.cos(3*x) +\
b_1 * np.sin(x) + b_2 * np.sin(2*x) + b_3 * np.sin(3*x)

plt.plot(x,f)
plt.show()
```
