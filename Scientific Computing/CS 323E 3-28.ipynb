```python
import numpy as np
from pylab import imshow, gray, show
    
def main():
    #define constants
    M = 100    #number of square on a side
    V = 1.0    #voltage at the top wall
    tol = 1e-3    #tolerance of solution
    
   # w = .01    #over-relaxation, arbitrary value to try to make code converge faster, may need different values
    
    #create arrays to hold potential values
    phi = np.zeros ([M+1,M+1], float)
    #initialize boundary condition
    phi[0,:] = V
    
    #main loop
    delta = 1.0   #arbitrary val, specify the difference
    while (delta > tol):
        maxDiff = 10000
        for i in range (M + 1):
            for j in range (M + 1):
                if (i == 0) or (i == M) or (j == 0) or (j == M):
                    phi[i,j] = phi[i,j]    #on the boundary, we do not change the values
                else:  # avg of 4 surrounding values
                    old = phi[i,j]
                    phi[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1]) / 4
                    newDiff = abs(phi[i,j] - old)
                    if newDiff < maxDiff:
                        maxDiff = newDiff
                
        #calculate max difference from old values
        delta = maxDiff
    
    #make a plot
    imshow (phi)
    gray()
    show()

main()
                
```
